```python
import ast 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
#loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
#print("All imports successful!")

#Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) # to change job skill list into data type
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA 
```

```python
df_DA['job_posted_month_no'] =df_DA['job_posted_date'].dt.month
```

```python
df_DA
```

```python
df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot = df_DA_pivot [df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_pivot = df_DA_pivot.drop('Total')
df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')

plt.title('Top 5 skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python

df_DA_pivot = df_DA_pivot.rename(columns={'index': 'job_posted_month_no'})

df_DA_pivot['job_posted_month_no'] = pd.to_numeric(df_DA_pivot['job_posted_month_no'], errors='coerce')

df_DA_pivot = df_DA_pivot[df_DA_pivot['job_posted_month_no'].between(1, 12)]

# Convert valid month numbers to abbreviated month names
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(
    lambda x: pd.to_datetime(str(int(x)), format='%m').strftime('%b')
)

# Set month names as the index
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')

# Drop the numeric month column
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

# Plot top 5 columns (skills)
df_DA_pivot.iloc[:, :5].plot(kind='line')
plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```
