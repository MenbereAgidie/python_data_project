```python
import ast 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
from matplotlib.ticker import FuncFormatter
import seaborn as sns
from matplotlib.ticker import PercentFormatter
#loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) # to change job skill list into data type
```

```python
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')] .copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month  #create a new column job_posted_month_no
```

```python
df_DA_US_expolode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_expolode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0 )

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)
```

```python
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')
```

```python
df_DA_US_percent
```

```python
skills = ['sql', 'excel', 'python', 'tableau', 'power bi']
df_plot = df_DA_US_percent[skills]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()
plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likehood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()

ax= plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i] )

plt.show()
```
